```python
import numpy as np
import time
import datetime
import numba
from math import lgamma, floor
import multiprocessing as mp
from numpy.random import Generator, PCG64, SeedSequence
from ill_confirm_bias_FNs_02f_cnt import confirm_bias_full
np.set_printoptions(suppress=True)

```

```python
start = time.perf_counter()

runs = 10
numarms = 2
maxalphabeta = 4
# numagents = 6
minagents = 10
maxagents = 17
pulls = 1000
payprobs = np.asarray([.499, .5])
tollerance = 0.
nsteps = 10
timesteps = 2*10**5

mem_depth = [5, 10, 100, 1000]
mem_snap = [10, 100, 1000, 10000] #these values must be multiples of nsteps to get recorded

erprobA = 0.5
erprobB = 2-2*(1-erprobA)**.5
print(f'erprobB = {erprobB}')

networktype = 'errandom'

mem = 50 # must be greater than max numagents if memshare = 0
memshare = 10
memstore = 5

pool = mp.Pool(processes=4)

for numagents in [4, 6, 9, 12, 15, 25]:
    sg = SeedSequence()
    rgs = [Generator(PCG64(s)) for s in sg.spawn(runs)]
    

    final_mp = pool.starmap(confirm_bias_full, [(n, mem_snap, mem_depth, erprobB, mem, memshare, memstore, networktype, maxalphabeta, pulls, numagents, numarms, tollerance, payprobs, nsteps, timesteps, rgs) for n in range(runs)])
        
#     final_mp = np.asarray(final_mp)
    suc_mp = [item[0] for item in final_mp]
    suc_mp = np.asarray(suc_mp)
    con_mp = [item[1] for item in final_mp]
    con_mp = np.asarray(con_mp)
    depth_mp = [item[2] for item in final_mp]
    depth_mp = np.asarray(depth_mp)
    
    depth_mean = np.mean(depth_mp, axis=0)
    
    sucrate = np.sum(suc_mp)/runs
    conrate = np.sum(con_mp)/runs
    
    print('time:')
    print(datetime.datetime.now())
    
    print(f'for # agents = {numagents} success rate = {sucrate} consensus rate = {conrate}')
    print('network average memory depths:')
    print(depth_mean)
    print(' ')
    print(' ')
    
finish = time.perf_counter()
print(f'Finished in {round(finish-start,0)/60} minutes')
```
